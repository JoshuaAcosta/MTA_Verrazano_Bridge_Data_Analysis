```python
import pandas as pd
```

```python
# Read in csv file showing the number of vehicles crossing MTA bridges and tunnels by E-Z pass subscribers and toll-by-mail
csv_file = "../data/MTA_Bridges_Tunnels_Hourly_Traffic_Rates__Beginning_2010_20241024.csv"
col_names = ['plaza_id', 'date', 'hour', 'direction', 'num_ezpass_vehicles', 'num_vtoll_vehicles']

df = pd.read_csv(csv_file, names=col_names, header=0)
df.shape
```

```python
df.head(10)
```

```python
#Filter plaza_id to only return Verrazano-Narrows Bridge
verrazano_df = df[df.plaza_id == 30]
verrazano_df.shape
```

```python
# Confirm no null values in dataframe
verrazano_df.isnull().sum()
```

```python
# Combine the date and hour column and extract date-related properties 
verrazano_df.loc[:,'datetime'] = pd.to_datetime(verrazano_df['date'] + ' ' + verrazano_df['hour'].astype(str).str.zfill(2) + ':00')
verrazano_df.loc[:,'day_of_week'] = verrazano_df['datetime'].dt.day_name()
verrazano_df.loc[:,'month'] = verrazano_df['datetime'].dt.month_name()
verrazano_df.loc[:,'year'] = verrazano_df['datetime'].dt.year

verrazano_df = verrazano_df[['plaza_id','datetime', 'day_of_week', 
                            'month', 'year', 'hour', 'direction',
                            'num_ezpass_vehicles','num_vtoll_vehicles']]


verrazano_df.head(10)
```

```python
# Replace values in direction volume for clarity
#OB - Staten Island; IB - Brooklyn

dir_values = {'I': 'Brooklyn', 'O':'Staten Island'}
verrazano_df['direction'] = verrazano_df.direction.replace(dir_values)
verrazano_df.head(5)
```
